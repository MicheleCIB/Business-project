Here we add two columns to our dataset: one containing market share (market share is the percent of total sales in an industry generated by a particular company.
```python
def click_counter(df1, df2, name='clicks'):
    clicks=[]
    id=list(pd.unique(df1['product_id']))
    for i in id:
        a=len(df2[df2['product_id']==i])
        clicks.append(a)
    df1[name]=clicks
    return df1
```

## TASK 1

To solve task 1, the leader and follower analysis for each period we decided to proceed like this: there are several variables that can be taken into account depending on the context and objectives of the analysis. Some of the common variables to consider include:

Market Share: the market share of a product or vendor can be an important indicator for identifying leaders and followers. A high market share might indicate the dominance of a leader, while a lower market share might suggest the presence of followers.

Sales volume: The sales volume of a product or seller can provide an indication of their popularity and success in the market. Leaders tend to have a higher sales volume than followers.

Price: The price of a product or service can influence its position in the market. Leaders may have higher prices than followers because of their reputation or the added value they offer. 

Price volatility can also be an indicator of the dynamism and instability of a market. Leaders might have lower price volatility, indicating greater stability and control over the market. On the other hand, followers might be subject to larger and more unpredictable fluctuations in price, indicating a position of less control or influence.

In our case, we decided that the leader could be the one who has either the highest price, minimal price volatility (less or equal to the mean of volatility for that period) or a fairly high market share (more than 15% of the total per period).

PERIOD 1

```python
dataset = period_data1
dataset['comp_date'] = pd.to_datetime(dataset['comp_date'])
first_day = dataset['comp_date'].min()
last_day = dataset['comp_date'].max()

results = pd.DataFrame(columns=['product_id', 'seller_id', 'leader_or_follower'])
 
unique_product_ids = dataset['product_id'].unique()
for product_id in unique_product_ids:
    product_data = dataset[dataset['product_id'] == product_id]
    total_market_share = product_data['market_share'].sum()
    max_price = product_data['price'].max()
    mean_volatility = product_data['volatility'].mean()

    is_leader = (max_price == product_data['price']) | (product_data['volatility'] <= mean_volatility) | (product_data['market_share'] > total_market_share+0.15)

    leaders = product_data[is_leader]
    followers = product_data[~is_leader]
    leaders['leader_or_follower'] = 'Leader'
    followers['leader_or_follower'] = 'Follower'

    results = results.append(leaders[['product_id', 'seller_id', 'leader_or_follower']])
    results = results.append(followers[['product_id', 'seller_id', 'leader_or_follower']])

results
```

Through this cycle, we print the graphs for each period and each product. On the x-axis we have the seller ids and the dot will be coloured red if follower and blue if leader. In our case, since we have set the Black Friday period, we will display a chart for each product (potentially up to more than 6000 charts depending on the number of different products), indicating whether or not there were leaders and followers.

```python
import matplotlib.pyplot as plt

data = results
product_ids = data['product_id'].unique()

for product_id in product_ids:
    product_data = data[data['product_id'] == product_id]
    seller_ids = product_data['seller_id']
    leader_or_follower = product_data['leader_or_follower']

    plt.figure(figsize=(15, 10))
    plt.scatter(seller_ids, [1] * len(seller_ids), c=leader_or_follower.map({'Leader': 'blue', 'Follower': 'red'}), cmap='coolwarm')
    plt.title(f'Leader vs Follower period 1 - Product ID: {product_id}')
    plt.xlabel('Seller ID')
    plt.ylabel('Status')

    legend_colors = {'Leader': 'blue', 'Follower': 'red'}
    plt.legend(handles=[plt.Line2D([], [], marker='o', color='w', markerfacecolor=color, markersize=8)
                        for label, color in legend_colors.items()],
               labels=legend_colors.keys(),
               loc='upper left', bbox_to_anchor=(1, 1))
    
    plt.show()
```

## TASK 2

We combine all the values using different weights for the variables. Then we add the new values to the respective dataframes. The first ones are for the q1, that doesn't have the number of bidding, the remaining ones are for the the remaining periods and for the Black Friday. 

We have chosen to determine the popularity index according to these criteria, using this weights;
Number of sales: represents the total number of units of a product sold during a certain time period. A high number of sales indicates a higher popularity of the product. It corresponds to quantity.

Online search: Indicates how often the product is searched for on search engines or e-commerce platforms. A high number of searches may indicate widespread interest in the product.

Profit: profit analysis can provide crucial information on the effectiveness of pricing strategies, the profitability of the product and its popularity in terms of revenue generation. An increase in profits may indicate an increase in demand or effective cost management, while a decrease in profits could be a sign of declining popularity or fiercer competition.

```python
# These weights are used to generate the column for the popularity index
a1=0.40 #quantity
b1=0.30 #number of regular clicks q1
c1=0.15 #profit for q1
d1=0.15 #totalstock

a=0.40 #quantity
b=0.25 #number of regular clicks
c=0.05 #number of bidding clicks
d=0.15 #profit
e=0.15 #totalstock
popindexq1['popularity']=popindexq1['quantity']*a1+popindexq1['clicks_regular']*b1+popindexq1['profit']*c1+popindexq1['total_stock']*d1
popindexq2['popularity']=popindexq2['quantity']*a+popindexq2['clicks_regular']*b+popindexq2['clicks_bidding']*c+popindexq2['profit']*d+popindexq2['total_stock']*e
popindexq3['popularity']=popindexq3['quantity']*a+popindexq3['clicks_regular']*b+popindexq3['clicks_bidding']*c+popindexq3['profit']*d+popindexq3['total_stock']*e
popindexq4['popularity']=popindexq4['quantity']*a+popindexq4['clicks_regular']*b+popindexq4['clicks_bidding']*c+popindexq4['profit']*d+popindexq4['total_stock']*e
popindexqnov['popularity']=popindexqnov['quantity']*a+popindexqnov['clicks_regular']*b+popindexqnov['clicks_bidding']*c+popindexqnov['profit']*d+popindexqnov['total_stock']*e
```

This function creates the graphs, in order to be able to detect the popularity index visualization.

```python
import seaborn as sn
def graph(popindexq1):
    plt.figure(figsize=(15, 10))
    sn.set_style("whitegrid")
    color = ['blue']
    g = sn.barplot(x="product_id", y="popularity", data=popindexq1.head(75), palette=color, order=popindexq1.head(75).sort_values('popularity', ascending = False).product_id)
    g.set_xticklabels(g.get_xticklabels(), rotation=50, horizontalalignment='right')
    g.set_ylabel('Popularity')
    g.set_xlabel('Product_id')
```
